Podemos ainda transformar a matriz com os valores absolutos em uma matriz de distribuiÃ§Ã£o de probabilidades

```python
def probability(matrix):
    matrix = matrix.astype(float)
    
    for line in matrix:
        maximum = sum(line)
        for i in range(len(line)):
            line[i] = line[i]/maximum
    
    return matrix
```

```python
df['matrix'] = df['matrix'].apply(lambda x: probability(x))
df['matrix'] = df['matrix'].apply(lambda x: x.transpose())
df
```

```python
import queue

"""A funÃ§Ã£o toma como argumento o conjunto de nucleotÃ­deos (a,c,g,t) em forma de matriz(4xn),
extrai os fatores de transcriÃ§Ã£o e retorna em forma de lista"""
def combinatoria (a, c, g, t):
    tam = len(a)
    ft = [''] # Lista de fatores de transcriÃ§Ã£o
    cont = 1 # Tamanho da lista de fatores de transcriÃ§Ã£o obs:tamanho mÃ­nimo==1
    q = queue.Queue() #fila auxiliar
    
    # Verifica se possui valor para o nucleotÃ­deo e adiciona na fila
    for i in range(tam):
        if a[i] != 0:
            q.put('A')
        if c[i] != 0:
            q.put('C')
        if g[i] != 0:
            q.put('G')
        if t[i]!=0:
            q.put('T')
        
        tam_fila = q.qsize() #Guarda o tamanho da fila
        indice=cont #Guarda o nÃºmero de iteraÃ§Ãµes que acontecerÃ¡ na lista de fatores de transcriÃ§Ã£o
        cont = cont*tam_fila #Aumenta o tamanho da lista de fatores de transcriÃ§Ã£o (cont==tamanho da lista de FT)
        
        # A combinaÃ§Ã£o sempre ocorre sobre todos os elementos da lista de fatores de transcriÃ§Ã£o
        if tam_fila>1: #Se o tamanho da fila for 1, nÃ£o Ã© necessÃ¡rio copiar os elementos para aumentar a lista de fatores de transcriÃ§Ã£o
            copy = ft #Faz uma cÃ³pia dos fatores de transcriÃ§Ã£o
            for j in range(tam_fila-1): #Copia os elementos para os novos Ã­ndices da lista
                ft = ft + copy
        
        #Extrai todos os nucleotÃ­deos da fila
        fila = []
        while not q.empty():
            for j in range(q.qsize()):
                fila.append(q.get())
                
        j=0 #variÃ¡vel auxiliar para o tamanho da lista de fatores de transcriÃ§Ã£o
        k=0 #variÃ¡vel auxiliar para interaÃ§Ã£o sobre a fila
        z=0 #variÃ¡vel auxiliar para interaÃ§Ã£o sobre a fila na lista de fatores de transcriÃ§Ã£o
        while (j < cont):
            ft[j] += fila[k]
            j = j+1
            z = z+1
            
            if z == indice:
                k = k+1
                z=0
        
    return ft
```

```python
"""
A funÃ§Ã£o toma como argumento uma matriz de distribuiÃ§Ã£o de probabilidades dos nucleotÃ­deos (a,c,g,t) 4xn 
e retorna uma Ãºnica string que representa todos os fatores de transcriÃ§Ã£o em forma de consenso pelo padrÃ£o IUPAC
"""
def degeneration (matrix):
    lst = []
    a = matrix[0,]
    c = matrix[1,]
    g = matrix[2,]
    t = matrix[3,]
    
    
    tam = len(a)
    i=0
    while i < tam:
        ia = 0
        ic = 0
        ig = 0
        it = 0
        if a[i] >= 0.3:
            ia = 1
        if c[i] >= 0.3:
            ic = 1
        if g[i] >= 0.3:
            ig = 1
        if t[i] >= 0.3:
            it = 1
        
        
        if ia + ic + ig + it == 4:
            lst.append ('N')
        elif ia + ic + ig + it == 1:
            if ia == 1:
                lst.append ('A')
            elif ic == 1:
                lst.append ('C')
            elif ig == 1:
                lst.append ('G')
            elif it == 1:
                lst.append ('T')
        
        elif ia + ic + ig + it == 3:
            if ia & ic & ig:
                lst.append ('V')
            elif ia & ic & it:
                lst.append ('H')
            elif ia & ig & it:
                lst.append ('D')
            elif ic & ig & it:
                lst.append ('B')
        
        elif ia + ic + ig + it == 2:
            if ia & ic:
                lst.append('M')
            elif ia & ig:
                lst.append('R')
            elif ia & it:
                lst.append('W')
            elif ic & ig:
                lst.append ('S')
            elif ic & it:
                lst.append ('Y')
            elif ig & it:
                lst.append('K')
        
        else:
            lst.append('N')
        
        
        i +=1
        
    return lst
```

```python
df['motifs'] = df.apply(lambda x: degeneration(x['matrix']), axis=1)
```

```python
#Conjunto de teste para mineraÃ§Ã£o de ECRs
promotor = pd.read_csv(
    '../promotores teste.txt',
    sep=',', header=None, names=['id','promotor'])
print(type(promotor), promotor.shape)
promotor
```

```python
"""
A funÃ§Ã£o toma como argumento uma regiÃ£o gÃªnica, denominada promotor, o tamanho dessa regiÃ£o
e um motivo (ECR) que serÃ¡ procurado dentro dessa regiÃ£o. A funÃ§Ã£o retorna uma lista com todos
os locais que o elemento Ã© encontrado.
"""
def searchFT (promotor, tamanho, motivo):
    promotor = promotor.lower()
    degeneracao = {
        'N' : ['a', 'c', 'g', 't'],
        'V' : ['a', 'c', 'g'],
        'H' : ['a', 'c', 't'],
        'D' : ['a', 'g', 't'],
        'B' : ['c', 'g', 't'],
        'M' : ['a', 'c'],
        'R' : ['a', 'g'],
        'W' : ['a', 't'],
        'S' : ['c', 'g'],
        'Y' : ['c', 't'],
        'K' : ['g', 't'],
        'A' : ['a'],
        'C' : ['c'],
        'G' : ['g'],
        'T' : ['t'],
    }
    
    lst = []
    tam = len(motivo)
    i = 0
    
    while (i < tamanho-tam-1):
        aminoacido = motivo[0]
        if (promotor[i] in degeneracao[aminoacido]):
            match = True
            j = 1
            while ((j < tam) & (match == True)):
                aminoacido = motivo[j]
                if (promotor[i+j] not in degeneracao[aminoacido]):
                    j=0
                    match = False
                    break
                else:
                    j += 1
            
            if ((match==True) & (j == tam)):
                lst.append(i)
        i += 1
        
    return lst
        
        
```

```python
"""
 A funÃ§Ã£o calcula o complemento reverso de um motivo para poder fazer a pesquisa do ECR atravÃ©s de
 uma Ãºnica fita de RNA.
"""
def reverseComplement(lst):
    tam = len(lst)
    i=0
    while (i < tam):
        if lst[i] == 'V':
            lst[i] = 'B'
        elif lst[i] == 'H':
            lst[i] = 'D'
        elif lst[i] == 'D':
            lst[i] = 'H'
        elif lst[i] == 'B':
            lst[i] = 'V'
        elif lst[i] == 'M':
            lst[i] = 'K'
        elif lst[i] == 'R':
            lst[i] = 'Y'
        elif lst[i] == 'Y':
            lst[i] = 'R'
        elif lst[i] == 'A':
            lst[i] = 'T'
        elif lst[i] == 'T':
            lst[i] = 'A'
        elif lst[i] == 'C':
            lst[i] = 'G'
        elif lst[i] == 'G':
            lst[i] = 'C'
        else:
            pass
        
        i += 1
    
    return lst
```

```python
#Cria uma coluna no data frame para o complemento reverso (a coluna sÃ³ recebe o motivo reverso)
df['reverseComplement'] = df['motifs'].apply(lambda x: x[::-1])
#Calcula o complemento do motivo (para resultar no complemento reverso)
df.reverseComplement.apply(lambda x: reverseComplement(x))
df
```

```python
def ecrMiner (promoter, ecr):
    searchLog = ecr.filter(['id', 'name', 'family_id', 'motifs', 'reverseComplement'], axis=1)
    searchLog['gene'] = ''
    searchLog['upstream'] = 0
    searchLog['downstream'] = 0
    searchLog['mean'] = 0
    searchLog['sum'] = 0
    searchLog

    for prom in promoter.itertuples():
        print (prom.locus)
        
        for motivo in ecr.itertuples():
            mean = 0.0
            positivo = searchFT (prom.promoter, prom.size, motivo.motifs)
            negativo = searchFT(prom.promoter, prom.size, motivo.reverseComplement)
            hits = len(positivo+negativo)
            if hits > 0:
                mean = sum(positivo+negativo)/hits
                searchLog = searchLog.append ({'gene':prom.locus, 
                                               'id':motivo.id,
                                               'name':motivo.name,
                                               'family':motivo.family_id,
                                               'motifs':motivo.motifs,
                                               'reverseComplement':motivo.reverseComplement,
                                               'upstream': positivo,
                                               'downstream':negativo,
                                               'mean':mean,
                                               'sum':hits}, ignore_index=True)
    
    return searchLog
```
