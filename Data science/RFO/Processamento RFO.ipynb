```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#Importando data frame da fatores de transcrição
df = pd.read_csv(
    '../df.csv',
    sep=',')
df = df.drop(['a', 'c', 'g', 't'], axis=1)
print(type(df), df.shape)
df
```

```python
#Importando data frame da fatores de transcrição
genoma = pd.read_csv(
    '../Genome Log.csv',
    sep=',')
genoma.drop(axis=1, columns=['name', 'motifs', 'reverseComplement'], inplace=True)
genoma
```

```python
df['vigna_genome'] = 0
genoma = genoma.drop(columns=['gene', 'upstream', 'downstream', 'mean'], axis=1)
i = genoma.groupby('id').sum()
df['vigna_genome'] = i['sum'].values.copy()
df
```

```python
#Importando data frame da analise de fenilpropanoides
drt1 = pd.read_csv(
    'Log RFO.csv',
    sep=',')
drt1
```

```python
drt1 = drt1.drop(columns=['name', 'gene', 'motifs', 'reverseComplement', 'upstream', 'downstream', 'mean'], axis=1)
df['Cluster'] = 0
i = drt1.groupby(['id']).sum()
df['Cluster'] = i['sum'].values.copy()
df
```

```python
df.to_csv(path_or_buf='Process Log RFO.csv',
          sep=',', header=True, index=True, mode='w')
```

```python
model = smf.ols("Cluster ~ name + vigna_genome",data=df)
result = model.fit()
```

```python
enrichment = pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
enrichment
```

```python
enrichment.to_csv(path_or_buf='Enrichment Log RFO.csv',
          sep=',', header=True, index=True, mode='w')
```
