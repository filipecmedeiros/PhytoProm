```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://phytoprom:ph1t0pr0m@localhost:5432/PhytoProm')
```

```python
"""
A função toma como argumento uma região gênica, denominada promotor, o tamanho dessa região
e um motivo (ECR) que será procurado dentro dessa região. A função retorna uma lista com todos
os locais que o elemento é encontrado.
"""
def searchFT (promotor, tamanho, motivo):
    promotor = promotor.lower()
    degeneracao = {
        'N' : ['a', 'c', 'g', 't'],
        'V' : ['a', 'c', 'g'],
        'H' : ['a', 'c', 't'],
        'D' : ['a', 'g', 't'],
        'B' : ['c', 'g', 't'],
        'M' : ['a', 'c'],
        'R' : ['a', 'g'],
        'W' : ['a', 't'],
        'S' : ['c', 'g'],
        'Y' : ['c', 't'],
        'K' : ['g', 't'],
        'A' : ['a'],
        'C' : ['c'],
        'G' : ['g'],
        'T' : ['t'],
    }
    
    lst = []
    tam = len(motivo)
    i = 0
    
    while (i < tamanho-tam-1):
        aminoacido = motivo[0]
        if (promotor[i] in degeneracao[aminoacido]):
            match = True
            j = 1
            while ((j < tam) & (match == True)):
                aminoacido = motivo[j]
                if (promotor[i+j] not in degeneracao[aminoacido]):
                    j=0
                    match = False
                    break
                else:
                    j += 1
            
            if ((match==True) & (j == tam)):
                lst.append(i)
        i += 1
        
    return lst
        
        
```

```python
def ecrMiner (promoter, ecr):
    searchLog = ecr.filter(['id', 'name', 'family_id', 'motifs', 'reverseComplement'], axis=1)
    searchLog['gene'] = ''
    searchLog['upstream'] = 0
    searchLog['downstream'] = 0
    searchLog['mean'] = 0
    searchLog['sum'] = 0
    searchLog

    for prom in promoter.itertuples():
        print (prom.locus)
        
        for motivo in ecr.itertuples():
            mean = 0.0
            positivo = searchFT (prom.promoter, prom.size, motivo.motifs)
            negativo = searchFT(prom.promoter, prom.size, motivo.reverseComplement)
            hits = len(positivo+negativo)
            if hits > 0:
                mean = sum(positivo+negativo)/hits
                searchLog = searchLog.append ({'gene':prom.locus, 
                                               'id':motivo.id,
                                               'name':motivo.name,
                                               'family':motivo.family_id,
                                               'motifs':motivo.motifs,
                                               'reverseComplement':motivo.reverseComplement,
                                               'upstream': positivo,
                                               'downstream':negativo,
                                               'mean':mean,
                                               'sum':hits}, ignore_index=True)
    
    return searchLog
```
